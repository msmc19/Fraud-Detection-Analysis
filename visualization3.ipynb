```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'cleaned_digital_wallet_transactions.csv'
df = pd.read_csv(file_path)

# Count transactions per location for each user
user_location_counts = df.groupby(['user_id', 'location']).size().unstack(fill_value=0)

# Select the top 20 users by transaction volume
top_users = user_location_counts.sum(axis=1).sort_values(ascending=False).head(20).index
top_user_location_data = user_location_counts.loc[top_users]

# Create a professional heatmap
plt.figure(figsize=(16, 10))
heatmap = sns.heatmap(
    top_user_location_data, 
    cmap='coolwarm', 
    annot=False, 
    cbar=True, 
    linewidths=0.5, 
    linecolor='white'
)

# Customize the heatmap for a professional look
heatmap.set_title('Transaction Volumes Across Locations for Top Users', fontsize=18, fontweight='bold', pad=15)
heatmap.set_xlabel('Location', fontsize=14, labelpad=10)
heatmap.set_ylabel('User ID', fontsize=14, labelpad=10)

# Adjust tick parameters for readability
heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=10, rotation=45, ha='right')
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=10)

# Finalize the layout
plt.tight_layout()

# Show the heatmap
plt.show()
```
