```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'cleaned_digital_wallet_transactions.csv'
data = pd.read_csv(file_path)

# Define the thresholds for high transaction amounts
threshold = data['product_amount'].quantile(0.90)  # 90th percentile
threshold_95 = data['product_amount'].quantile(0.95)  # 95th percentile
threshold_99 = data['product_amount'].quantile(0.99)  # 99th percentile

# Create a new column flagging high transaction amounts
data['high_transaction_flag'] = data['product_amount'] > threshold

# Set a professional style using seaborn
sns.set_theme(style="whitegrid")

# Plot a histogram with thresholds using seaborn
plt.figure(figsize=(12, 8))
sns.histplot(data['product_amount'], bins=50, kde=False, color='skyblue', edgecolor='black', alpha=0.7, label='All Transactions')
plt.axvline(threshold, color='blue', linestyle='dashed', linewidth=2, label=f'90th Percentile (${threshold:.2f})')
plt.axvline(threshold_95, color='green', linestyle='dashed', linewidth=2, label=f'95th Percentile (${threshold_95:.2f})')
plt.axvline(threshold_99, color='orange', linestyle='dashed', linewidth=2, label=f'99th Percentile (${threshold_99:.2f})')

# Enhance title and axis labels
plt.title('Distribution of Transaction Amounts with Fraud Thresholds', fontsize=16, fontweight='bold')
plt.xlabel('Transaction Amount ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Add legend with better positioning
plt.legend(fontsize=12, loc='upper right')

# Customize grid and axes
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlim(left=0)  # Ensure x-axis starts at 0

# Final layout adjustments
plt.tight_layout()

# Display the improved plot
plt.show()
```
